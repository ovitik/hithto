# Recursive Thought Tokens Demo

Tiny tensor-only demo; it validates the latent hierarchy without downloading Gemma.

```python
import torch
from thought_tokens import ThoughtBuilder, ThoughtBuilderConfig

builder = ThoughtBuilder(ThoughtBuilderConfig(hidden_size=32, grouping='fixed', fixed_window=3, max_levels=2, max_group_size=4))
hidden = torch.randn(1, 10, 32)
out = builder(hidden)
len(out.all_thoughts), out.embeddings.shape, out.compression_ratios
```
